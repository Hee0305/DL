## k-means


```python
from sklearn.datasets import make_blobs

# 1.data load
X, y = make_blobs(n_samples=150,
                  n_features=2,
                  centers=3,
                  cluster_std=0.5,
                  shuffle=True,
                  random_state=42)

import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1],
            c='white', marker='o', edgecolor='black', s=50)
plt.grid()
plt.tight_layout()

from sklearn.cluster import KMeans
km = KMeans(n_clusters=3,
            init='random', #k-means++
            n_init=10,
            max_iter=300,
            tol=1e-04,
            random_state=42)
y_km = km.fit_predict(X)


plt.scatter(X[y_km == 0,0],
            X[y_km == 0,1],
            s=50, c='lightgreen',
            marker='s', edgecolor='black',
            label='Cluster 1')


plt.scatter(X[y_km == 1,0],
            X[y_km == 1,1],
            s=50, c='orange',
            marker='o', edgecolor='black',
            label='Cluster 2')


plt.scatter(X[y_km == 2,0],
            X[y_km == 2,1],
            s=50, c='lightblue',
            marker='v', edgecolor='black',
            label='Cluster 3')

plt.scatter(km.cluster_centers_[:,0],
            km.cluster_centers_[:,1],
            s=250, marker='*',
            c='red', edgecolor='black',
            label='Centroids')

plt.legend(scatterpoints=1)
plt.show()  

# ---------------------------- 엘보우 방법을 사용하여 최적의 클러스터 개수 찾기 ------------------------- #
print('왜곡: %.2f' % km.inertia_)

distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i,
                init='k-means++',
                n_init=10,
                max_iter=300,
                random_state=42)
    km.fit(X)
    distortions.append(km.inertia_)
plt.plot(range(1,11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()
```

덴드로그램

```python
import pandas as pd
import numpy as np 

# 데이터프레임 생성
np.random.seed(123)
variables = ['X', 'Y', 'Z']
labels = ['ID_0', 'ID_1','ID_2','ID_3','ID_4']
X = np.random.random_sample([5,3])*10
df = pd.DataFrame(X, columns=variables, index=labels)
print(df)


# 데이터셋 모든 샘플 쌍의 유클리디안 거리 계산
from scipy.spatial.distance import pdist, squareform 

row_dist = pd.DataFrame(squareform(
                        pdist(df, metric='euclidean')),
                        columns=labels, index=labels)
print(row_dist)

# 군집 결과
from scipy.cluster.hierarchy import linkage

row_clusters = linkage(df.values,
                       method='complete',
                       metric='euclidean')

'''
# 군집 결과를 df 변환
pd.DataFrame(row_clusters, 
             columns=['row label 1',
                      'row label 2',
                      'distance',
                      'no. of items in clust'],
             index=['cluster %d' %(i+1) for i in range(row_clusters.shape[0])])
'''

# 덴드로그램 그리기 
from scipy.cluster.hierarchy import dendrogram 
import matplotlib.pyplot as plt
row_dendr = dendrogram(row_clusters,
                       labels=labels)

plt.tight_layout()
plt.ylabel('Euclidean distance')
plt.show()

```

```python
# -*- coding: utf-8 -*-
"""
Created on Tue Sep 28 12:52:11 2021

@author: admin
"""

from sklearn import datasets
# import numpy as np


## 1. data load
iris = datasets.load_iris()

X = iris.data
y = iris.target

X = X[:,1:3]

#print('class label: ', np.unique(y))


## 2. abnormal data check (NaN, outlier check!!!)



## 3. train / test data split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = \
            train_test_split(X,y,random_state=42,test_size=0.2,stratify=y)
            
# print(np.bincount(y))
# print(np.bincount(y_train))
# print(np.bincount(y_test))



# ## 4. feature scaling
# from sklearn.preprocessing import StandardScaler
# ss = StandardScaler()
# ss.fit(X_tr)
# X_tr_std = ss.transform(X_tr)
# X_val_std = ss.transform(X_val)





## 5. algorithm training
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski')



## 6. performance measure
## stratified cross-validation 사용법
from sklearn.model_selection import cross_val_score

scores = cross_val_score(estimator = knn, 
                         X = X_train, 
                         y = y_train,
                         cv = 10,
                         n_jobs = -1)

print('CV accuracy(all): %s' %scores)




import numpy as np

print('CV accuracy: %.3f +/- %.3f' %(np.mean(scores), np.std(scores)))






mean_accuracy, nK_val = [], []
n_k = [1,3,5,7,9]
for k in n_k:
    knn2 = KNeighborsClassifier(n_neighbors=k, p=2, metric='minkowski')
    scores = cross_val_score(estimator = knn2, 
                             X = X_train, 
                             y = y_train,
                             cv = 10,
                             n_jobs = -1)
    nK_val.append(k)
    mean_accuracy.append(np.mean(scores))
    
print(mean_accuracy)



idx = mean_accuracy.index(max(mean_accuracy))
print("best k values: %.3f" %n_k[idx])
knn_opt = KNeighborsClassifier(n_neighbors=n_k[idx], p=2, metric='minkowski')


knn_opt.fit(X_train, y_train)    
y_test_pred = knn_opt.predict(X_test)

from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, f1_score

print("test data accuracy: %.3f" %accuracy_score(y_test, y_test_pred))

# y_test_prob = knn_opt.predict_proba(X_test_std)



## specificity (특이도)



## NPV



print("test data recall: %.3f" %recall_score(y_test, y_test_pred, average='weighted'))
print("test data precison: %.3f" %precision_score(y_test, y_test_pred, average='weighted'))
print("test data f1 score: %.3f" %f1_score(y_test, y_test_pred, average='weighted'))
# print("test data AUC: %.3f" %roc_auc_score(y_test, y_test_prob, multi_class='ovr'))
# print("test data Confusion matrix:")
print(confusion_matrix(y_test, y_test_pred))
    
    
    



```
